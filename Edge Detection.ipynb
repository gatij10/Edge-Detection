```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=plt.imread('yourname.jpg')
n,m,d=img.shape
edges_img=img.copy()

plt.imshow(edges_img)
plt.show()
```

```python
#Horizontal edge detection
horizontal_filter = [[-1,0,1],[-2,0,2],[-1,0,1]]
for row in range(3,n-2):
    for col in range(3,m-2):
        local_pixels=img[row-1:row+2,col-1:col+2,0] #little 3x3 box
        
        horizontal_transformed_pixels=horizontal_filter*local_pixels #Applying horizontal filter
        horizontal_score=horizontal_transformed_pixels.sum()/4
        
        edge_score=(horizontal_score**2 )**.5
        edges_img[row,col]=[edge_score]*3
        
edge_img=edges_img/edges_img.max()
plt.imshow(edge_img)
plt.show()
```

```python
#Vertical edge detection
vertical_filter = [[-1,-2,-1],[0,0,0],[1,2,1]]
for row in range(3,n-2):
    for col in range(3,m-2):
        local_pixels=img[row-1:row+2,col-1:col+2,0] #little 3x3 box
        
        vertical_transformed_pixels=vertical_filter*local_pixels #Applying vertical filter
        vertical_score=vertical_transformed_pixels.sum()/4
        
        edge_score=(vertical_score**2 )**.5
        edges_img[row,col]=[edge_score]*3
        
edge_img=edges_img/edges_img.max()
plt.imshow(edge_img)
plt.show()
```
